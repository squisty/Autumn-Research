```python
import numpy as np
import matplotlib.pyplot as plt
import math
import matplotlib.ticker as ticker


file = open('table2.dat')
```

```python
lst = []
```

```python
table = np.genfromtxt("table2.dat",delimiter=" ",names=["xpix","ypix","vmag","vmag error","b-v","b-v error"])
vmag = table["vmag"]
```

```python
for line in file:
    lst += [line.split()]
```

```python
vmag = [x[3] for x in lst]
vmaga = np.array(vmag)
print(type(vmaga[0]))
```

```python
bv = [x[5] for x in lst]
bva = np.array(bv)
```

```python
fig, ax = plt.subplots(figsize=(13,13))
ax.scatter(bva, vmaga, s=5, alpha = .5)
tick_spacing = 100
xmax, xmin = max(bva), min(bva)
ymax, ymin = max(vmaga), min(vmaga)

ax.invert_yaxis()
ax.set(xlabel='B-V',
    ylabel='Mv',)
ax.set_title('B-V vs. Mv')
ax.set_yticklabels([])
#ax = fig.add_axes([xmin, xmax])
#start, end = ax.get_xlim()
#ax.xaxis.set_ticks(np.arange(start, end))
ax.set_xticklabels([])
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
#plt.show()
plt.savefig('/Users/samquist/Documents/GitHub/Autumn-Research/badcurve.png')





```

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.stats import LombScargle

def curvefunc(file):
    
    
    hdu = fits.open(file)
    data = hdu[1].data
    ok = np.where((data['QUALITY'] == 0) & (data['FCOR'] > 0))
    plt.figure(1)
    plt.scatter(data['TIME'][ok], data['FCOR'][ok], s = 5, alpha = 0.5)
    plt.ylim(7800,7930)
    
    
    LS = LombScargle(data['TIME'][ok], data['FCOR'][ok])
    frequency, power = LS.autopower(minimum_frequency=1./40., maximum_frequency=1./0.1)
    print("Maximum power = " + str( max(power)))
    plt.figure(2)
    plt.plot(1/frequency, power)
    plt.xscale('log')
    
    best_frequency = frequency[np.argmax(power)]
    print("Estimated Period = " + str(1/best_frequency) + " days")
    phase_fit = np.linspace(0, 1)
    y_fit = LombScargle(data['TIME'][ok], data['FCOR'][ok]).model(t=phase_fit / best_frequency,
                                        frequency=best_frequency)
    phase = (data['TIME'][ok] * best_frequency) % 1

    fig, ax = plt.subplots(figsize=(10, 4.5))
    ax.errorbar(phase, data['FCOR'][ok],yerr= data['FRAW_ERR'][ok], 
            fmt='o', mew=0, capsize=0, elinewidth=1.5, alpha = .5)
    ax.plot(phase_fit, y_fit, color='r')
    ax.invert_yaxis()
    ax.set(xlabel='phase',
        ylabel='magnitude',
        title='phased data at frequency={0:f}'.format(best_frequency))
    
    return

curvefunc('/Users/samquist/Downloads/hlsp_everest_k2_llc_220626248-c08_kepler_v2.0_lc.fits')
```
